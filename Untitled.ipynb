```python
FEATURES = [
    "OverallQual", "GrLivArea", "GarageCars", "TotalBsmtSF", "1stFlrSF",
    "FullBath", "TotRmsAbvGrd", "YearBuilt", "YearRemodAdd", "Fireplaces",
    "LotArea", "MasVnrArea", "BsmtFinSF1", "GarageArea", "LotFrontage", "WoodDeckSF",  
    "Neighborhood", "BldgType", "HouseStyle", "ExterQual", "KitchenQual",
]

CAT_COLS =["Neighborhood", "BldgType", "HouseStyle", "ExterQual", "KitchenQual",]
```

```python
df4 = df3[FEATURES + ["SalePrice"]].copy()
df4.head()
```

```python
for col in df4.select_dtypes(include=[np.number]).columns:
    df4[col] = df4[col].fillna(df4[col].median())
```

```python
from scipy import stats

z_scores = stats.zscore(df4[['GrLivArea', 'SalePrice']])
df5 = df4[(abs(z_scores) < 3).all(axis=1)].copy()
df5.shape
```

```python
from sklearn.preprocessing import LabelEncoder 
le_dict = {}
for col in CAT_COLS:
    le = LabelEncoder()
    df5[col] = le.fit_transform(df5[col].astype(str))
    le_dict[col] = list(le.classes_)
df5.head()
```

```python
X = df5.drop('SalePrice', axis = 'columns')
y = df5.SalePrice
y.head()
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingRegressor
import numpy as np

gbm = GradientBoostingRegressor(random_state=42)
param_dist = {
    "n_estimators": np.arange(100, 500, 50),
    "learning_rate": np.linspace(0.01, 0.2, 10),
    "max_depth": [2, 3, 4, 5, 6],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

random_search = RandomizedSearchCV(
    estimator=gbm,
    param_distributions=param_dist,
    n_iter=20,         
    cv=5,
    scoring="r2",
    random_state=42,
    n_jobs=-1
)

random_search.fit(X, y)
print("Best Params:", random_search.best_params_)
print("Best Score:", random_search.best_score_)
```
